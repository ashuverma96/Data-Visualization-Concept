```python
#plotting using matplotlib
##plt scatter

plt.scatter(x,y,c='g')
plt.xlabel('x axis')
plt.xlabel('y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
## plt plot

plt.plot(x,y,'r*--',linestyle='dashed',linewidth=2, markersize=12)
```

```python
## Creating subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
x = np.arange(1,11)
y = 3* x+5
plt.title("Matplotlib dia")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y)
plt.show
```

```python
# Compute the x and y coordinates for points on sine curve

x = np.arange(0,4 * np.pi, 0.1)
y = np.sin(x)
plt.title("sine wave form")

# Plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
#Subplot
#Compute the x and y coordinates for poitns on sine and cosine curve

x = np.arange(0,5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.

plt.subplot(2,1,1)

#Make the first plot

plt.plt(x,y_sin)
plt.title('Sine')

#Set the second subplot as active, and make the second plot

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine')

#Show the figure

plt.show()
```

```python
# Bar Plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x,y)
plt.bar(x2,y2, color = 'g')
plt.title("Bar Graph")
plt.ylabel('Y axis ')
plt.xlabel('X axis')

plt.show()
```

## Histogram

```python
a = np.array([22,87,5,43,56,55,54,11,20,51,5,79,31,27])
plt.hist(a, bins=20)
plt.hist(a)
plt.title("histogram")
plt.show()
```

## Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]

#Rectangular box plot
plt.boxplot(data, vert=True, patch_artist=True);
```

## Pie Chart

```python
# Data to plot
labels = 'Python','C++','Ruby','Java'
size = [215,130,245,210]
colors = ['gold', 'yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0) #explode 1st slice

#Plot 
plt.pie(size, explode=explode, labels=labels, colors=colors, autopct = '%1.1f%%',shadow=True)

plt.axis('equal')
plt.show()
```
