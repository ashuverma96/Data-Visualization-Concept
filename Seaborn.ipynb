## Seaborn

## Distribution Plot

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

## Correlation with Heatmap

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## JoinPlot


## Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

## Pair plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

## Dist plot

Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=True,bins=10)
```

## Categorical Plots

```python
## Count plot

sns.countplot('day',data=df)
```

```python
# Bar Plot
sns.barplot(x='total_bill',y='sex',data=df)
```

## Box Plot

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,palette='rainbow')
```

```python
# Categorize my data based on some other categories
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

## Violin Plot 

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
